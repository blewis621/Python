# Sampling Data

```python
# Create a list holding true/false values to test np.random.rand()<= 0.7
np.random.seed(4)
dev = np.random.rand(len(df_ords_prods_merge)) <= 0.7
```

```python
dev
```

```python
# Store 70 % of the sample in dataframe big
big = df_ords_prods_merge[dev]
```

```python
# Store 30 % of the sample in dataframe small
small = df_ords_prods_merge[~dev]
```

```python
# Checking dataframe size
len(df_ords_prods_merge)
```

```python
# Checking dataframe size
len(big) + len(small)
```
