# Data Consistency Checks

```python
# Describe dataframe
df_customers.describe()
```

```python
# Check for mixed-type data
for col in df_customers.columns.tolist():
  weird = (df_customers[[col]].applymap(type) != df_customers[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df_customers[weird]) > 0:
    print (col)
```

```python
# Change data type of First Name to string
df_customers['First Name'] = df_customers['First Name'].astype('str')
```

```python
# Find missing values
df_customers.isnull().sum()
```

There are no missing values in the customers columns of the dataset.

```python
# Looking for duplicates
df_customers_dups = df_customers[df_customers.duplicated()]
```

```python
df_customers_dups
```

There are no duplicate values within the customer dataset.
